```python
# 利用决策树做回归问题的分析
#!/urs/bin/python
# -*- coding:utf-8 -*-

import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor

# 使得负号正常显示
mpl.rcParams['axes.unicode_minus'] = False

def main():
    # 定义N个数据点
    N = 100
    # 随机生成N个数据点
    x = np.random.rand(N) * 6 - 3     #[-3, 3)
    # 对x 进行排序
    x.sort()
    y = np.sin(x) + np.random.randn(N) * 0.1
#     print(y)
#     print('x:\n', x)
    x = x.reshape(-1, 1)
#     print('x:\n', x)
#     model = DecisionTreeRegressor(criterion='mse', max_depth=10)
    # n_estimators=20 表示一个随机森林中 决策树的个数，这里是20个
    model = RandomForestRegressor(n_estimators=20, criterion='mse', max_depth=10)
    model.fit(x, y)
    x_test = np.linspace(-3, 3, 50).reshape(-1, 1)
    y_hat = model.predict(x_test)
    
    plt.figure(facecolor='w')
    plt.plot(x, y, 'r*', markersize=15, markeredgecolor='k', label='实际值')
    plt.plot(x_test, y_hat, 'gv', linewidth=2, label='预测值')
    plt.legend(loc='upper left', fontsize=12)
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.grid(b=True, ls=':', color='#606060')
    plt.title('决策树的随机森林回归', fontsize=15)
    plt.tight_layout(2)
    plt.show()
    
    depth = [2, 4, 6, 8, 10]
    clr = 'rgbmy'
#     model = DecisionTreeRegressor(criterion='mse')
    model = RandomForestRegressor(n_estimators=20, criterion='mse')
    plt.figure(facecolor='w')
    plt.plot(x, y, 'ro', ms=5, mec='k', label='实际值')
    x_test = np.linspace(-3, 3, 50).reshape(-1, 1)
    for d, c in zip(depth, clr):
        model.set_params(max_depth=d)
        model.fit(x, y)
        y_hat = model.predict(x_test)
        plt.plot(x_test, y_hat, '-', color=c, linewidth=2, markeredgecolor='k', label='Depth=%d'%d)
    plt.legend(loc='upper left', fontsize=12)
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.grid(b=True, ls=':', color='#606060')
    plt.title('决策树/随机森林回归', fontsize=15)
    plt.tight_layout(2)
    plt.show()
    
    
    
    print('\nDone!')
    
if __name__ == "__main__":
    main()
```
