```python
# 
#!/urs/bin/python
# -*- coding:utf-8 -*-

import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor

# 
mpl.rcParams['axes.unicode_minus'] = False

def main():
    # N
    N = 100
    # N
    x = np.random.rand(N) * 6 - 3     #[-3, 3)
    # x 
    x.sort()
    y = np.sin(x) + np.random.randn(N) * 0.1
#     print(y)
#     print('x:\n', x)
    x = x.reshape(-1, 1)
#     print('x:\n', x)
#     model = DecisionTreeRegressor(criterion='mse', max_depth=10)
    # n_estimators=20  20
    model = RandomForestRegressor(n_estimators=20, criterion='mse', max_depth=10)
    model.fit(x, y)
    x_test = np.linspace(-3, 3, 50).reshape(-1, 1)
    y_hat = model.predict(x_test)
    
    plt.figure(facecolor='w')
    plt.plot(x, y, 'r*', markersize=15, markeredgecolor='k', label='')
    plt.plot(x_test, y_hat, 'gv', linewidth=2, label='')
    plt.legend(loc='upper left', fontsize=12)
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.grid(b=True, ls=':', color='#606060')
    plt.title('', fontsize=15)
    plt.tight_layout(2)
    plt.show()
    
    depth = [2, 4, 6, 8, 10]
    clr = 'rgbmy'
#     model = DecisionTreeRegressor(criterion='mse')
    model = RandomForestRegressor(n_estimators=20, criterion='mse')
    plt.figure(facecolor='w')
    plt.plot(x, y, 'ro', ms=5, mec='k', label='')
    x_test = np.linspace(-3, 3, 50).reshape(-1, 1)
    for d, c in zip(depth, clr):
        model.set_params(max_depth=d)
        model.fit(x, y)
        y_hat = model.predict(x_test)
        plt.plot(x_test, y_hat, '-', color=c, linewidth=2, markeredgecolor='k', label='Depth=%d'%d)
    plt.legend(loc='upper left', fontsize=12)
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.grid(b=True, ls=':', color='#606060')
    plt.title('/', fontsize=15)
    plt.tight_layout(2)
    plt.show()
    
    
    
    print('\nDone!')
    
if __name__ == "__main__":
    main()
```
