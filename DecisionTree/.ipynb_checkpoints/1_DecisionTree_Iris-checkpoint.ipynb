```python
# 基于鸢尾花数据进行的决策树代码实现
#!/urs/bin/python
# -*- coding:utf-8 -*-

import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
# sklearn中模型预处理包sklearn.preprocessing
from sklearn.preprocessing import StandardScaler
# 导入决策树分类器 DecisionTreeClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree



# 定义鸢尾花数据类型转换函数
def iris_type(s):
    # 定义转换字典
    # u 后面的字符串，以 Unicode 格式 进行编码，一般用在中文字符串前面
    # r 后面的字符串，声明是普通字符串，使其包含的转义字符（反斜杠）失效
    # b前缀代表的就是bytes 
    it = {b'Iris-setosa': 0, b'Iris-versicolor' : 1, b'Iris-virginica' : 2}
    return it[s]

def main():
    # 实现中文显示
#     mpl.rcParams['font.sans-serif'] = ['SimHei']
#     mpl.rcParams['axes.unicode_minus'] = False
    
    # 定义鸢尾花数据特征
    iris_feature_E = 'sepal length', 'sepal width', 'petal length', 'petal width'
    iris_feature = u'花萼长度', u'花萼宽度', u'花瓣长度', u'花瓣宽度'
    # 定义鸢尾花类别
    iris_class = 'Iris-setosa', 'Iris-versicolor', 'Iris-virginica'
    # 定义数据文件路径
    path = './iris.data'
    
#     # 利用pandas读取数据，得到DataFrame类型或者Series类型
#     data = pd.read_csv(path, header=None)
#     print(type(data))
#     print(data.head())
#     print(data.tail())
#     print(list(range(4)))    # 定义读取数据列的列表
#     # 读取data的特征数据，并赋给 x
#     x = data[list(range(4))]

    # 利用numpy读取数据
    # converters={4:iris_type} 是转换器，使其第4列按照 iris_type函数定义的字典转换
    data = np.loadtxt(path, dtype=float, delimiter=',', converters={4 : iris_type})
    print(type(data))
#     print(data)
    # 分离读取的原始数据，赋值给 x, y    其中(4,) 的含义为从索引为4的位置切片
    x, y = np.split(data, (4,), axis=1)
    print('x-type: ', type(x), '\ny-type: ', type(y))
#     print(x, y)
    # 为了可视化，仅取x的前两列特征
    x = x[:, :2]    # 左闭右开
    # 定义训练数据和测试数据
    # train_test_split()对数据集随机划分成训练数据和测试数据，train_size=0.7表示70%是训练数据
    # random_state=1 定义随机数种子为1
    x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.7, random_state=1)
    
    # 决策树估计
    # 利用管道 Pipeline 定义决策树模型,Pipeline()的参数通常是列表
    model = Pipeline([
        # 进行特征缩放，进行标准化，保证每一个特征的均值是0，方差是1；以便提高分类效果
        ('ss', StandardScaler()),
        # criterion='entropy' 表示使用 entropy 生成决策树，max_depth=3 最大深度为3（不包括根节点）
        ('DTC', DecisionTreeClassifier(criterion='entropy', max_depth=3))
        
    ])

#     # 利用一般形式定义决策时模型，而不使用Pipeline，相互等价
#     # 定义StandardScaler实例 ss
#     ss = StandardScaler()
#     # fit_transform()的作用就是先拟合数据，然后转化它将其转化为标准形式
#     # 调用 StandardScaler的实例 ss 对数据进行标准化
#     x = ss.fit_transform(x)
#     # 定义决策树分类器
#     model = DecisionTreeClassifier(criterion='entropy', max_depth=3)
    
    
    # 模型的训练
    model.fit(x_train, y_train)
    # 利用模型对测试数据进行预测
    y_test_hat = model.predict(x_test)
    # 保存生成的决策树模型的可视化文件 iris_tree.dot
    # 打开 iris_tree.dot  如果没有该文件，则创建一个这样的文件
    f = open('./iris_tree_20190718.dot', 'w')
    # 调用tree的export_graphviz函数，把模型输出到可视化文件 out_file
    tree.export_graphviz(model.get_params('DTC')['DTC'], out_file=f)
#     # 如果不适用管道 Pipeline,则适用下面一行代码保存模型
#     tree.export_graphviz(model, out_file=f)
    f.close()
    
    
    # 画图——使用决策树对数据集处理的可视化
    # 定义横x0 纵x1 坐标轴的采样点数目
    N, M = 100, 100
    
    # 定义第0 列的范围    
    x0_min, x0_max = x[:, 0].min(), x[:, 0].max()
    # 定义第1 列的范围
    x1_min, x1_max = x[:, 1].min(), x[:, 1].max()
    
    # 在x0 轴上采样N 个点，即100个点 (一维数据)    t0: (100, )     [...]
    t0 = np.linspace(x0_min, x0_max, N)    
    # 同理    t1: (200, )
    t1 = np.linspace(x1_min, x1_max, M)
#     print('t0:\n', t0)
#     print('t1:\n', t1)
#     print('t0 shape: ', t0.shape)
#     print('t1 shape: ', t1.shape)

    # 生成网格采样点t0 => x0 : (N,) => [M, N]     t1 => x1 : (M, ) => [M, N]
    # meshgrid(t0, t1)  表示以 t0为第1维度（axis=0），t1为第2维度（axis=1），生成数据
    x0, x1 = np.meshgrid(t0, t1)
#     print('x0:\n', x0)
#     print('x1:\n', x1)
#     print('x0 shape: ', x0.shape)
#     print('x1 shape: ', x1.shape)

    # 把x0, x1 打平成一维
    x0_flat, x1_flat = x0.flat, x1.flat
#     print('x0:\n', x0)
#     print('x1:\n', x1)
    
    # 把x0, x1  在axis=1，即列的方向上堆叠起来，形成一个个的二维坐标点,即测试点
    x_show = np.stack((x0_flat, x1_flat), axis=1)
#     print('x_show type: ', type(x_show))
#     print('x_show shape: ', x_show.shape)    # [100*100, 2]
#     print('x_show:\n', x_show)
    
    # 定义分类的背景颜色图例（较浅）
    cm_light = mpl.colors.ListedColormap(['#A0FFA0', '#FFA0A0', '#A0A0FF'])
#     print('cm_light type: ', type(cm_light))
#     print('cm_light:\n', cm_light)
    # 定义样本点的颜色类（较深）
    cm_dart = mpl.colors.ListedColormap(['g', 'r', 'b'])
    
    # 对生成的坐标点进行测试，以便使用生成的决策树对坐标系平面划分分类区域，
    y_show_hat = model.predict(x_show)
#     print('x1 shape: ', x1.shape)
#     print('x_show shape: ', x_show.shape)
#     print('y_show_hat shape: ', y_show_hat.shape)
    # 把 y_show_hat 的形状变成可以在坐标系里标志的形状   (10000,) => [100, 100]
    y_show_hat = y_show_hat.reshape(x1.shape)
#     print('y_show_hat reshape: ', y_show_hat.shape)
#     print('y_show_hat:\n', y_show_hat)
    
    # 开始绘制可视化图
    # 定义画板，其画板颜色是白色的 'w'
    plt.figure(facecolor='w')
    # 使用cm_light中的颜色，对 y_show_hat 进行绘制,（在x0 和 x1 轴所表征的坐标平面绘制）
    plt.pcolormesh(x0, x1, y_show_hat, cmap=cm_light)
    
    # 对ravel() 函数的考究
#     print(y_test)
#     print('y_test shape: ', y_test.shape)    # [45, 1]
#     print(y_test.ravel())
#     print('y_test shape: ', y_test.ravel().shape)    #(45,)

    # 绘制样本点(散点)
    '''
    绘制测试样本点
    c=y_test.ravel() 表示：定义选用颜色，使用y_test的对应的颜色，调用ravel()使得y_test成为一行，函数要求
                            需要行（一维的）
    edgecolors='k'   表示：定义绘制散点的边 的颜色：'k’表示黑色
    s=100            表示：定义绘制散点的面积
    cmap=cm_dart     表示：定义绘制散点的颜色
    marker='o'       表示：定义绘制散点的形状
    '''  
    plt.scatter(x_test[:, 0], x_test[:, 1], c=y_test.ravel(), edgecolors='k', s=100, cmap=cm_dart,
               marker='*')
    # 绘制全部样本点
    plt.scatter(x[:, 0], x[:, 1], c=y.ravel(), edgecolors='k', s=30, cmap=cm_dart)
    
    # 定义 x， y 轴的标签
    plt.xlabel(iris_feature[0], fontsize=15)
    plt.ylabel(iris_feature[1], fontsize=15)
    
    # 定义 x, y 轴的取值范围，不留白
    plt.xlim(x0_min, x0_max)
    plt.ylim(x1_min, x1_max)
    
    # 定义坐标图中网格线
    #  linestyle='--' 定义线形 -- 表示虚线
    plt.grid(True, linestyle='--')
    
    # 定义标题   u'....'带不带 u 无所谓
    plt.title(u'鸢尾花数据的决策树分类', fontsize=17)
    
    # 显示命令
    plt.show()
    
    
    # 计算在测试集上的预测结果，即准确度
    # 使y_test 成为一维的  (45, 1)  => (45,)
#     print('y_test shape: ', y_test.shape)
    y_test = y_test.reshape(-1)
#     print('y_test shape: ', y_test.shape)
#     print('y_test: ', y_test)
#     print('y_test_hat: ', y_test_hat)
    # 比较 y_test 与 y_test_hat
    result = (y_test_hat == y_test)
#     print(result)    # [ True  False ......   True] True——1 预测正确；False——0 预测错误
    # 计算正确率
    acc = np.mean(result)
    print('准确度: %.3f%%'%(100*acc))    # %%  表示要输出一个%
    
    # 描绘出不同的层数，相应的准确度
    # 层数： 1 ~ 15
    depth = np.arange(1, 15)
    # 定义错误率列表
    err_list = []
    for d in depth:
        # 调用sklearn决策树模型
        clf = DecisionTreeClassifier(criterion='entropy', max_depth=d)
        clf = clf.fit(x_train, y_train)
        # 在测试集上测试模型
        y_test_hat = clf.predict(x_test)
        result = (y_test == y_test_hat)
        err = 1 - np.mean(result)
        err_list.append(err)
        print('决策树为 %d 层时，准确度为: %.3f%%.'%(d, 100*np.mean(result)))
        
    # 画图表示决策树深度与过拟合的关系
    plt.figure('w')
    plt.plot(depth, err_list, 'ro-', lw=1)
    plt.xlabel('决策树的深度', fontsize=15)
    plt.ylabel('错误率', fontsize=15)
    plt.title('决策树深度与过拟合的关系', fontsize=17)
    plt.grid(True, linestyle='--')
    plt.show()
    
    print('\n\nDone!')
    pass


if __name__ == "__main__":
    main()
```
